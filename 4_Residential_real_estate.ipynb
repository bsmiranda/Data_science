## Residential Real Estate

### In the previous file (3_Types_of_real_estate) an analysis was made of what were the types of properties contained in the database. Here, an analysis was made to filter only residential properties from the database, removing other types of properties that we were not interested in, such as Industrial, Commercial Complex, Standard Land, etc ... After filtering only the types of properties that we are interested in for analysis, a new file was created in csv format using the to_csv function, this file contains a database with the residential properties that interest us.

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head(10)
```

```python
# list with the types of residence contained in the Type column
list(dados['Tipo'].drop_duplicates())
```

```python
# I just need the types of residential properties
# that's why I created the 'residential' variable, which stores residential properties
residencial = [ 'Quitinete',
                'Casa',
                'Apartamento',
                'Casa de Condomínio',
                'Casa de Vila']
```

```python
residencial
```

```python
dados.head(10)
```

```python
# Isin checks if the values are contained in the Series and returns Boolean values
# returns True on the lines where the properties selected as residential are
# returns False on the lines where non-residential properties are contained
selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
# creating a new variable that stores the selection data, we now have only residential properties
# there are 22,580 residential properties out of 32,960 types of properties in total
dados_residencial = dados[selecao]
dados_residencial
```

```python
# here are the objects on the list, containing only the types of properties we want
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
dados_residencial.shape[0]
```

```python
# the index function displays the first 10 items and the last 10 items in a list
# we can sort with the range function
dados_residencial.index
```

```python
# sorting with the range function
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial
```

## Exporting the Database

```python
# exporting the file with the to_csv function with the separator (';')
# we now have a document in csv format in our database
dados_residencial.to_csv('dados/aluguel_residencial.csv', sep = ';')
```
